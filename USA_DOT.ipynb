```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

Reading in the excel data and merging the sheets into one dataframe with category info attached to row data
    (the way the data was organized in the excel doc made this quite difficult!)

```python
# read in data, skip rows so that headings become column names
xls = pd.ExcelFile('in_US_1960to2018_class.xlsx')
```

```python
# read in data, set the header to be the desired column titles
df = pd.read_excel(xls, header = 3, nrows= 7)
df # view the data frame
```

```python
# rename the class categories by adding a column
df['class']=['all','M1','L','N1','N2','N3','M2']
```

```python
# drop original class descriptors
df.drop(columns='Unnamed: 0', inplace= True)
```

```python
# melt dataframe to bring years into rows.
melted_df= df.melt(id_vars='class', var_name= 'year')
nomissingvalues= melted_df[melted_df['value']!= 'U']
nomissingvalues.head(4)
```

## structuring the data into format of datastructure


```python
# rename the dataframe to df
df= nomissingvalues
df
```

```python
# add in a column of indexes
df.index = 'FHA' + df.index.astype(str)
```

```python
#rename or add all necessary columns
df.loc[:,'year_of_measurement']= df.loc[:,'year']
df.loc[:,'date_of_measurement']=''
df.loc[:,'geo']= 'US'
df.loc[:,'process']= 'in_use'
df.loc[:,'vehicle_class']= df['class']
df.loc[:,'vehicle_segment']='all'
df.loc[:,'drive_train']= 'all'
df.loc[:,'model_year']= 'all'
df.loc[:,'year_of_first_registraion']=''
df.loc[:,'value']= df['value']
df.loc[:,'unit']= 'nr'
df.loc[:,'source']= 'FHA'
df.loc[:,'accessed']= '2020-06-28'
df.loc[:,'notebook']= 'USA_DOT'
df.loc[:,'footnote']= 'c'

heading_list= ['year_of_measurement','date_of_measurement','geo','process','vehicle_class','vehicle_segment','drive_train','model_year','year_of_first_registraion','value','unit','source','accessed','notebook','footnote']
df_out= df[heading_list]

df_out.head(1)
```

```python
df_out.to_excel('out_DOT_US_st_1993_to_2018.xlsx', merge_cells=False)
```
